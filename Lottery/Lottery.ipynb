```python
import random
prize = set()

while len(prize) < 7:
    n = random.randint(1, 48)
    prize.add(n)
print("Prize number：", prize)
```

```python
times = 0
while True:
    lottery = set()
    while len(lottery) < 7:
        n = random.randint(1, 48)
        lottery.add(n)
    print("My lottery：", lottery)
    times = times + 1
    
    total = 0
    for n in lottery:
        if n in prize:
            print(n, "中了")
            total = total + 1
    print("中了", total,"個數字")
    
    if total >=5:
        break
print("共買了", times,"張才中獎")    
```
